#### Count Or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
#pre processing
#dropping the first column
train_set.drop(0,axis=1,inplace=True) #axis indicates column
```

```python
train_set
```

```python
columns=[1,3,5,6,7,8,9,13]
train_set=train_set[columns]
```

```python
train_set.head()
```

```python
#add new colunm names:
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head(15)
```

```python
train_set["Employment"].unique().__len__()
```

```python
for i in train_set.columns[:]:
    print(i,",",len(train_set[i].unique()),"Labels")
```

```python
county_map=train_set["Country"].value_counts().to_dict()
county_map
# convert it into dictionary//if it have same values like germany and canada have 127 ,we cant do
#we can add mokup data  or cook up data
```

```python
train_set["Country"]=train_set["Country"].map(county_map)
#mapped with frequency
```

```python
train_set["Country"].value_counts()
```

#### Target Guided Ordinal Encoding
1. Ordering the labels according to the target(output)
2. Replace the labels by the joint probability of being 1 or 0

```python
tit=pd.read_csv("titanic.csv",usecols=["Cabin","Survived"])
```

```python
tit.head()
```

```python
tit["Cabin"].fillna("Missing",inplace=True)
```

```python
tit["Cabin"].unique()
#c = floor // room number :46 ,6
#we are taking the floor i,e : 85,46,6,78
```

```python
tit["Cabin"].astype(str).str[0].unique()
# we requied the floor 
#cabin is in object type
#we use astype to convert to string
#then cut string from object type
#str[0]:first index of the string 
```

```python
tit['Cabin'] = tit['Cabin'].astype(str).str[0]
```

```python
# how many ppl escapes from each cabin 
#to rank :: probability number, of ppl escapes from each cabin 
#find the highest ppl escapes from each cabin 
#we need to take the highest ppl escapes from mean 

tit.Cabin.unique()
```

```python
o_label=tit.groupby(['Cabin'])['Survived'].mean().sort_values().index
#we need to take the highest ppl escapes from mean 
```

```python
mean_o  #mean encoding based on the target value
```
