```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[20,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("day",fontsize=10)#for label x:axis and y:axis
plt.ylabel("temp")
plt.title("weather info:::")
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
plt.show()
```

```python
#BAR CHART
company=["google","FB","TATA","IOC"]
revenue=[29,45,67,21]
profit=[11,34,7,12]

```

```python
plt.xlabel("companies")
plt.ylabel("rev")
plt.title("company & revenue")
plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label="profit")
plt.legend()#stacking graph
plt.show()
```

```python
len(company)
```

```python
xposition=np.arange(len(company))
xposition
```

```python
plt.bar(xposition,revenue)
```

```python
plt.title("company & revenue")
plt.bar(xposition,revenue,label="revenue")
plt.bar(xposition,profit,label="profit")
plt.legend()#stacking graph
plt.show()
```

```python
xposition+0.9
```

```python
plt.xticks(xposition,company)
plt.title("company & revenue")
plt.bar(xposition-0.2,revenue,width=0.4,label="revenue")
plt.bar(xposition+0.2,profit,width=0.4,label="profit")
plt.legend()#stacking graph
plt.show()
```

```python
plt.title("company & revenue")
plt.barh(xposition,revenue,label="revenue")
plt.barh(xposition,profit,label="profit")
plt.legend()#stacking graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plot
```

```python
np.random.seed(10)
data=np.random.normal(100,20,200)#100=mean,20=sd,200=rows
```

```python
data
```

```python
plt.hist(data,rwidth=0.8)#about 70% of data lies b/w 80-120,one step SD // we convert into histogram to knw frequency distribution 
#20 step should be sd
#normal distribution follows a bell curve
```

```python
len(data)
```

```python
plt.boxplot(data)
plt.figure(figsize= (30,5))
plt.show()#the line inside the box is the median,mean=100 {if the data set follows normal distribution :mean ,sd ,mode are equal }
```

```python
np.mean(data)
```

```python
np.median(data)
```

```python
np.random.seed(10)
data1=np.random.normal(100,10,200)
data2=np.random.normal(90,20,200)
data3=np.random.normal(80,30,200)
data4=np.random.normal(70,40,200)
dat=[data1,data2,data3,data4]
plt.figure(figsize= (10,10))
plt.boxplot(dat,vert=False,patch_artist=True)
plt.show()
```

```python
values = [1400,600,300,410,250]
exp_labels = ["rent","food","phone","car","others"]
plt.figure(figsize = (10,7))#it should be the 1st line
plt.pie(values,labels= exp_labels,radius=2,autopct="%0.2f%%",shadow=True,explode=[0.2,0,0,0.2,0],startangle=90)#radius=to increase the size,[default=1],[SHAWDOW:: IS THE BLUE LINE AROUND THE PIE CHART]
plt.legend()
#plt.figure(figsize = (10,7))
plt.axis("equal")#auto fit
#plt.show()
plt.savefig("piechart1.jpg")
```
